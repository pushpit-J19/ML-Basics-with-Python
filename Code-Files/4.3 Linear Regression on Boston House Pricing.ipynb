```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
x = boston.data
y = boston.target
```

```python
x.shape, y.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2)
```

```python
xtrain.shape, ytrain.shape, xtest.shape, ytest.shape
```

```python
lr_model = LinearRegression(normalize=True)
lr_model
```

```python
lr_model.fit(xtrain, ytrain)
```

```python
lr_model.coef_
# can see there about 13 variables are dependent here for target
# so 13 m or slopes or dimensions
# we can visualize for 3d this is 13d
```

```python
lr_model.intercept_
# intercept with the target axis
```

## Accuracy

R2 SCORE =1 - (ya - fi)^2 / ((ya - yamean)^2)

ya = actual data

fi = predicted value

yamean = mean of actual y of training

-----------------

R^2 = 1 - {RSS} / {TSS}

The sum of squares of residuals, also called the residual sum of squares:

RSS = sum (y-f)^2

The total sum of squares (proportional to the variance of the data):

TSS = sum (y - ymean)^2


```python
lr_model.score(xtrain, ytrain)
```

```python
lr_model.score(xtest, ytest)
```

```python
lr_model.predict(xtest) # gives the predicted values based on fitted model
# can run a loop to calculate RSS and TSS using ytest
# to calculate R^2 accuracy
```
