```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def drawImg (img, title="Image"):
  #plt.axis(emit=False)
  plt.imshow(img, cmap="gray")
```

image.png

```python
img = plt.imread("/content/dog.jpg")
```

```python
drawImg(img), img.shape
# 900 rows, 900 columns, 3 channels : RGB 
# so converting to a 2d matrix
```

```python
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img1 = cv2.resize(img1, (100,100))
drawImg(img1), img1.shape
```

```python
def convolution (img, img_filter):
  W = img.shape[1]   # width
  H = img.shape[0]   # height
  
  F = img_filter.shape[0]   # filter is a matrix 3x3
  # print(W,H,F)
  new_image = np.zeros((H-F+1, W-F+1))  #98, 98  
  for row in range(0, H-F+1):   
    for col in range(0, W-F+1):
      # traversing the image
      for i in range(F):  
        for j in range(F):
          # traversing the Kernel
          new_image[row][col] += img[row+i][col+j]*img_filter[i][j]
        if new_image[row][col] > 255:
          new_image[row][col] = 255
        elif new_image[row][col] < 0 : 
          new_image[row][col] = 0

  return new_image
```

### BLUR

```python
# blur 
# decrease the pixel value
#blur_filter = np.ones((10,10)) / 9.0
blur_filter = np.ones((3,3)) / 9.0
print(blur_filter)
```

```python
output_img = convolution(img2, blur_filter)
drawImg(output_img)
```

```python
output_img = convolution(img1, blur_filter)
drawImg(output_img)
```

## Edge

```python
# Edge Filter
edge_filter = np.array([[1,0,-1],
                       [1,0,-1],
                       [1,0,-1]])
  
```

```python
output_edge = convolution(img2, edge_filter)
drawImg(output_edge)
```
